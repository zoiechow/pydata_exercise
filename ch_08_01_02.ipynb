```python
# pandas.merge可根据一个或多个键将不同DataFrame中的行连接起来。


# pandas.concat可以沿着一条轴将多个对象堆叠到一起。

# 实例方法combine_first可以将重复数据编接在一起，
# 用一个对象中的值填充另一个对象中的缺失值。
```

```python
df1 =pd.DataFrame({'key':['b','b','c','d','e','a','a'],'data1':range(7)})
df1
```

```python
df2 = pd.DataFrame({'data2':range(3),'key':['a','b','c']})
df2
```

```python
df1.merge(df2)
```

```python
df2.merge(df1)
```

```python
pd.merge(df1,df2)
```

```python
# 注意，我并没有指明要用哪个列进行连接。如果没有指定，
# merge就会将重叠列的列名当做键。不过，最好明确指定一下：
pd.merge(df2,df1,on = 'key')
```

```python
# 如果两个对象的列名不同，也可以分别进行指定：
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                  'data2': range(3)})
df4
```

```python
pd.merge(df3,df4,left_on ='lkey',right_on = 'rkey')
# left 与right必须成对出现
```

```python
# 外连接求取的是键的并集，组合了左连接和右连接的效果：
pd.merge(df3,df4,left_on ='lkey',right_on = 'rkey',how = 'left')
# how 的属性有inner，outer，left，right
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                     'data1': range(6)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
df2
```

```python
pd.merge(df1,df2,on = 'key')
# 多对多连接产生的是行的笛卡尔积。
# 由于左边的DataFrame有3个"b"行，右边的有2个，所以最终结果中就有6个"b"行。
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                      'lval': [1, 2, 3]})
left
```

```python
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
right
```

```python
pd.merge(left,right,on = ['key1','key2'])
```

```python
pd.merge(left,right,on = ['key1','key2'],how = 'inner')
```

```python
pd.merge(left,right,on = ['key1','key2'],how = 'outer')
```

```python
pd.merge(left,right,on = 'key1',how = 'inner')
```
