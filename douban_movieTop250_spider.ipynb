```python

import pandas as pd
import numpy as np
import json
```

```python
from collections import defaultdict
data = defaultdict(list)
for i in range(10):
    url = 'https://movie.douban.com/top250?start={}&filter='.format(i*25)
    con = requests.get(url).content
    sel = html.fromstring(con)
    for i in sel.xpath('//div[@class = "info"]'):
        title = i.xpath('div[@class = "hd"]/a/span[@class = "title"]/text()')[0]
        info = i.xpath('div[@class ="bd"]/p[1]/text()')
        star = i.xpath('''div[@class = "bd"]/div[@class = "star"]
                       /span[@class = "rating_num"]/text()''')[0]
    # satr = i.xpath('//span[@calss = "rating_num"]/text()')[0]
        num_com = i.xpath('div[@class = "bd"]/div[@class = "star"]/span/text()')[1]
    # num_com = i.xpath('//div[@class = "star"]/span[4]/text()') [0]


    #   信息获取
    #   导演信息
#         director = info[0].strip().replace("\n","").split("主演")[0]
        
#         actor = "主演" + info[0].strip().replace("\n","").split("主演")[1]

        director = info[0].strip().replace("\n","")
    
    
        date = info[1].strip().replace("\n","").split("/")[0]

        country = info[1].strip().replace("\n","").split("/")[1]

        geners = info[1].strip().replace("\n","").split("/")[2]



#         print(title,director,date,country,geners,star,num_com)
        data['title'].append(title)
        data['country'].append(country)
        data['date'].append(date)
        data['geners'].append(geners)
        data['num_com'].append(num_com)
        data['star'].append(star)
        data['director'].append(director)
```

```python
data = pd.DataFrame(data)
```
