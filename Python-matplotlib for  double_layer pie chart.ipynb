```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vals = [1,2,3,4]#创建数据系列 
```

```python
fig ,ax  = plt.subplots()#创建子图 
```

```python
labels = 'a','b','c','d'
```

```python
colors = ['yellow','red','blue','pink']
```

```python
explode = (0,0.1,0,0)
```

```python
ax.pie(vals,explode = explode,labels = labels ,colors = colors,autopct='%1.1f%%',shadow = True,startangle =90,radius = 1.2)
ax.set(aspect = "equal",title = 'pie plot with ax.pie')
plt.show(ax.pie)
```

```python
vals = [1, 2, 3, 4]#创建数据系列 
fig, ax = plt.subplots()#创建子图 
labels = 'A', 'B', 'C', 'D' 
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'] 
explode = (0, 0.1, 0, 0) 
ax.pie(vals, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90,radius=1.2) 
ax.set(aspect="equal", title='Pie plot with `ax.pie`')#设置标题以及图形的对称 
plt.show()
# 某一块凸出（explode）进行设置后，最终呈现的图形效果。
# pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6,
# shadow=False, labeldistance=1.1, startangle=None, radius=None)
# pie（）函数中有个饼图半径的设置——radius
```

```python
var1 = [2,3,4,5]
var2 = [7,8,9,10]
fig,ax = plt.subplots()
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'] 
labels = 'a','b','c','d'
ax.pie(var1,radius = 2,autopct='%1.1f%%',shadow = True,pctdistance = 0.9,colors = colors)
ax.pie(var2,radius = 1.5,autopct='%1.2f%%',pctdistance = 0.5,shadow = True)
# 对pctdistance（百分比离圆形距离）进行不同设置，以便区分开
# autopct='%1.2f%%' 数据百分比的小数数量
ax.set(aspect="equal", title='Pie plot with `ax.pie`') 
plt.legend(labels,bbox_to_anchor = (3,1),loc = 'best' ,borderaxespad = 3)
# 图例的展示设置 ，bbox_to_anchor = (3,1）距离图形的位置多远
plt.show()
# va1的半径要大于第二个，不然会被遮盖
```
